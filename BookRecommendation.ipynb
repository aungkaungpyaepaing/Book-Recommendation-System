*This code first groups the final_rating DataFrame by User-ID and Book-Title, and calculates the mean rating for each group using the mean() method. The result is a DataFrame with three columns: User-ID, Book-Title, and Book-Rating*

```python
# Aggregate the ratings by taking the mean rating for each book by each user
mean_ratings = final_rating.groupby(['User-ID', 'Book-Title'])['Book-Rating'].mean().reset_index()

# Create a User-Book matrix with NaN values
matrix = mean_ratings.pivot(index='User-ID', columns='Book-Title', values='Book-Rating')

# Print the shape of the matrix
print('Shape of User-Book matrix:', matrix.shape)

# Print the number of non-NaN values in the matrix
print('Number of non-NaN values in matrix:', matrix.count().sum())

matrix.head()
```

# Date Normalization

```python
matrix_nom = matrix.subtract(matrix.mean(axis=1), axis = 'rows')
matrix_nom
```

# User Similarity

There are different way to measure user similarity. Pearson correlation and Cosine similarity are two widely use methods. Here, had used Pearson correlation to calculate the user similarity matrix. 

If want to use cosine, have to change NaN to 0. 

```python
#user similarity using Pearson correlation
user_similarity = matrix.T.corr()
user_similarity
```

Finding user similarity example

Need to exclude User ID from the similar user list and decide the number of similar user

```python
#Pick User ID
picked_user = 16

#remove pick user id from the candidate list 
user_similarity.drop(index=picked_user, inplace = True)

user_similarity.head(10)
```

In user similarity matrix, the value of range is from -1 to 1.
-1 mean opposide preference and 1 mean same preference

# Picking at least top 10 similar user 

n = 10 mean number of user would like to pick
user similarity threshold need to be positive since like similarity, meaning that user must have a Perason Correlation coefficent of at least 0.3 to be consider as a similar user

Sort from highest to low, then print similar user id

```python
n = 10

#user similarity threshold
user_similarity_threshold = 0.3

#getting top n similar users
similar_users = user_similarity[user_similarity[picked_user]>user_similarity_threshold][picked_user].sort_values(ascending=False)[:n]

#print out top n similar users
print(f'The similar user {picked_user} are ', similar_users)
```

# Book Recommendation System

To make book recommendation system, have to remove all the books that had been read by selected user and keep only the book that the similar users had read

**Step 1**
To remove the book that the user had read, keep the role for user=id and remove the missing value

```python
picked_user_read = matrix[matrix.index == picked_user].dropna(axis=1, how='all')
picked_user_read
```

**Step 2**

Books that the similar user had read

```python
similar_user_books = matrix[matrix.index.isin(similar_users.index)].dropna(axis=1, how='all')
similar_user_books
```

**Step 3**
Removing books that selected user had read

```python
similar_user_books.drop(picked_user_read.columns,axis=1, inplace=True, errors='ignore')

similar_user_books
```

# Deciding Which books to recommend



```python
# A dictionary to store item scores
item_score = {}
# Loop through items
for i in similar_user_books.columns:
  # Get the ratings for book i
  book_rating = similar_user_books[i]
  # Create a variable to store the score
  total = 0
  # Create a variable to store the number of scores
  count = 0
  # Loop through similar users
  for u in similar_users.index:
    # If the book has rating
    if pd.isna(book_rating[u]) == False:
      # Score is the sum of user similarity score multiply by the rating
      score = similar_users[u] * book_rating[u]
      # Add the score to the total score for the book so far
      total += score
      # Add 1 to the count
      count +=1
  # Get the average score for the item
  item_score[i] = total / count
# Convert dictionary to pandas dataframe
item_score = pd.DataFrame(item_score.items(), columns=['Book', 'Book_Score'])
    
# Sort the book by score
ranked_item_score = item_score.sort_values(by='Book_Score', ascending=False)
# Select top book
m = 10
ranked_item_score.head(m)
```

```python
from surprise import SVD
from surprise import Dataset, Reader
from surprise.model_selection import cross_validate
from surprise.model_selection import train_test_split
import numpy as np
from surprise import accuracy


# Load the data using the Surprise library
reader = Reader(rating_scale=(1, 10))
data = Dataset.load_from_df(final_rating[['User-ID', 'ISBN', 'Book-Rating']], reader)

# Split the data into training and test sets
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

# Train FunkSVD model
funk_svd_model = SVD(n_factors=50, biased=True, verbose=True)
funk_svd_model.fit(trainset)

# Make predictions on the test set and calculate RMSE
funk_svd_preds = funk_svd_model.test(testset)
funk_svd_accuracy = accuracy.rmse(funk_svd_preds)

# Make predictions on the test set and calculate MAE
funk_svd_preds = funk_svd_model.test(testset)
funk_svd_accuracy = accuracy.mae(funk_svd_preds)

# Get top-N recommendations for each user using FunkSVD model
from collections import defaultdict

def get_top_n(predictions, n=10):
    # First map the predictions to each user.
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))
    
    # Then sort the predictions for each user and retrieve the n highest ones.
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]
    
    return top_n

top_n = get_top_n(funk_svd_preds, n=10)
```
